1. Top 3 and bottom 3 cities by total trips over the 6 month period

```python
import pandas as pd

# Aggregating total trips by city
city_trip_summary = factTrips.groupby('city_id')['trip_id'].count().reset_index()
city_trip_summary.rename(columns={'trip_id': 'total_trips'}, inplace=True)

# Sorting cities by total trips
city_trip_summary_sorted = city_trip_summary.sort_values(by='total_trips', ascending=False)

# Top 3 and Bottom 3 performing cities
top_3_cities = city_trip_summary_sorted.head(3)
bottom_3_cities = city_trip_summary_sorted.tail(3)

# Merging with city names for better readability
top_3_cities = top_3_cities.merge(dimCity, on='city_id', how='left')
bottom_3_cities = bottom_3_cities.merge(dimCity, on='city_id', how='left')

print("Top 3 Performing Cities:")
print(top_3_cities[['city_name', 'total_trips']])

print("\nBottom 3 Performing Cities:")
print(bottom_3_cities[['city_name', 'total_trips']])
```

- Observations
- Jaipur, Lucknow and Surat had the highest number of trips 
- Visakhapatnam, Coimbatore and Mysore had the least number of trips

2. Average Fare Per Trip, Average Distance(in km) per trip and Fare per kilometer

```python
#Calculate average fare per trip by city
city_fare_summary = factTrips.groupby('city_id').agg(
    total_fare= ('fare_amount','sum'),
    total_trips= ('trip_id','count'),
    total_distance = ('distance_travelled(km)','sum'),
).reset_index()

city_fare_summary['avg_fare_per_trip'] = city_fare_summary['total_fare']/ city_fare_summary['total_trips']
city_fare_summary['avg_fare_per_trip'] 
city_fare_summary['avg_dist_per_trip'] = city_fare_summary['total_distance'] / city_fare_summary['total_trips']
city_fare_summary = city_fare_summary.merge(dimCity,on='city_id', how='left')

fare_per_kilometer= city_fare_summary['avg_fare_per_trip']/ city_fare_summary['avg_dist_per_trip']
fare_per_kilometer
city_fare_summary['fare_per_kilometer'] = fare_per_kilometer
#city_fare_summary
#--- Rearranging the columns
city_fare_summary = city_fare_summary[['city_id','city_name','total_distance','total_fare','total_trips','avg_fare_per_trip','avg_dist_per_trip','fare_per_kilometer']]
city_fare_summary
```

```python
# Sort by Cities with highest fare per kilometer to lowest fare per kilometer.
city_fare_summary = city_fare_summary.sort_values(by='fare_per_kilometer', ascending=False)
city_fare_summary
```

Observations:-
- Jaipur and Mysore are the cities with the highest fare per kilometer
- Surat and Vadodra are the citis with the lowest fare per kilometer

6. Repeat Passengers Analysis
- Analyze the frequency of trips taken by the repeat passengers in each city (eg, % of repeat passengers taking 2 trips, 3 trips etc,). Identify which cities contribute most to higher trip frequencies amoong repeat passsengers. 

```python
print (repeatTrip.groupby(['city_id','trip_count']).sum().reset_index())
```

```python
# Calculate the total number of repeat passengers for each city
city_total_repeat_passengers = repeatTrip.groupby('city_id')['repeat_passenger_count'].sum().reset_index()
city_total_repeat_passengers.rename(columns={'repeat_passenger_count': 'total_repeat_passengers'}, inplace=True)

# Merge the total repeat passengers back with the repeatTrip dataset
repeatTrip_with_totals = repeatTrip.merge(city_total_repeat_passengers, on='city_id', how='left')

# Calculate the percentage of repeat passengers for each trip frequency in each city
repeatTrip_with_totals['percentage'] = (
    repeatTrip_with_totals['repeat_passenger_count'] / repeatTrip_with_totals['total_repeat_passengers'] * 100
)

# Aggregate data to find which cities contribute most to higher trip frequencies
# Filter for higher trip frequencies (e.g., 5 trips or more)
high_frequency = repeatTrip_with_totals[repeatTrip_with_totals['trip_count'].str.extract(r'(\d+)').astype(int) >= 5]
high_frequency_summary = high_frequency.groupby('city_id')['repeat_passenger_count'].sum().reset_index()
high_frequency_summary = high_frequency_summary.merge(dimCity, on='city_id', how='left')

# Sort by the highest contribution to high frequencies
high_frequency_summary = high_frequency_summary.sort_values('repeat_passenger_count', ascending=False)

# Display the results
print("Percentage of Repeat Passengers by Trip Frequency (City-Wise):")
print(repeatTrip_with_totals[['city_id', 'trip_count', 'percentage']])

print("\nCities Contributing Most to Higher Trip Frequencies (5 trips or more):")
print(high_frequency_summary[['city_name', 'repeat_passenger_count']])
```
